```
library(ggplot2) #安裝並產生範例用的畫圖資料

df<-data.frame(gp=factor(rep(letters[1:3],each=10)),y=rnorm(30))

ggplot(df, aes(x = gp, y = y)) +geom_boxplot() #畫盒狀圖
```

```
df$z<-df$y+rnorm(30)

ggplot(df,aes(x=z,y=y))+geom_line()+facet_grid(gp~.) #畫線，畫折線圖
```
